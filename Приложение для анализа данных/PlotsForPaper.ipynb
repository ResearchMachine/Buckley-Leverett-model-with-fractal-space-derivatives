```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


Variable = 'Накопленная доб. за 180 дней'
sns.set(font_scale=2)

WellName = [134, 201, 202, 206, 212, 214, 223]


IP90fact = np.array([899.5, 2543.8, 3310.5, 2090.4, 3478.5, 3864.9, 4480.8])
IP90prog = np.array([924.5064, 2418.8080, 3285.52, 2179.8665, 3522.5712, 3797.7621, 4539.4225])

IP180fact = np.array([2669, 5061, 5648, 4980, 5685, 7265, 8158])
IP180prog = np.array([2683.78574, 4987.04989, 5633.214252, 5032.929292, 5711.046475, 7225.271964, 8192.702374])


fig, ax = plt.subplots(figsize=(14,8))
ax.set_xlabel('Факт')
ax.set_ylabel('Прогноз')
sns.regplot(IP90fact, IP90prog, color ='blue', ci=100)
sns.regplot(IP180fact, IP180prog, color ='red', ci=100)
ax.legend(['Накоп. добыча за 90 дней', 'Накоп. добыча за 180 дней'])
```
